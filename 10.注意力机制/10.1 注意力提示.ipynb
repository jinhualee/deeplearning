## 10.1.2 查询、键和值

自主性的与非自主性的注意力提示解释了人类的注意力的方式，
下面我们看看如何通过这两种注意力提示，
用神经网络来设计注意力机制的框架，

首先，**考虑一个相对简单的状况，
即只使用非自主性提示。
要想将选择偏向于感官输入，
我们可以简单地使用参数化的全连接层，
甚至是非参数化的最大池化层或平均池化层**。

因此，“是否包含自主性提示”将注意力机制与全连接层或池化层区别开来。
在注意力机制的背景下，我们将自主性提示称为**查询（query）**。
给定任何查询，注意力机制通过**注意力池化（attention pooling）**
将选择引导至**感官输入（sensory inputs，例如中间特征表示）**。
在注意力机制中，这些感官输入被称为**值（value）**。
更通俗的解释，每个**值**都与一个**键（key）** 配对 ，
这可以想象为感官输入的非自主提示。
如图10.1.3所示，我们可以设计注意力池化，
以便给定的查询（自主性提示,  Volitional Cues）可以与键（非自主性提示, Non-volitional Cues）进行匹配，
这将引导得出最匹配的值（感官输入）。

鉴于上面所提框架在图10.1.3 中的主导地位，
因此这个框架下的模型将成为本章的中心。
然而，注意力机制的设计有许多替代方案。
例如，我们可以设计一个不可微的注意力模型，
该模型可以使用强化学习方法进行训练。

- **要点：**
  - 在设计神经网络的注意力机制时，我们可以使用自主性和非自主性的注意力提示。
  - 非自主性的提示可以通过使用参数化的全连接层或者非参数化的最大池化层或平均池化层来实现。
  - 在注意力机制中，自主性提示被称为“查询”，而感官输入被称为“值”。每个“值”都与一个“键”（可以理解为感官输入的非自主性提示）配对。
  - 注意力机制通过注意力池化，将给定的查询（自主性提示）与键（非自主性提示）进行匹配，从而引导选择最匹配的值（感官输入）。
  - 除了这种框架，还有其他的注意力机制设计，如不可微的注意力模型，该模型可以使用强化学习方法进行训练。

- **说明：**
- - **（1）如何理解：“首先，考虑一个相对简单的状况， 即只使用非自主性提示。 要想将选择偏向于感官输入， 我们可以简单地使用参数化的全连接层， 甚至是非参数化的最大池化层或平均池化层”？**
  - 在注意力机制的语境中，非自主性提示（也被称作“自下而上”或“外显”注意力提示）通常是由**外部刺激驱动**的，这意味着对注意力的分配是由外部感官输入的特征决定的，不需要任何明显的**意图**或**控制**。例如，在视觉场景中，非自主性注意力可能会被不寻常、显眼或突出的物体所抓取。
  - 全连接层和池化层在神经网络中常用于从输入数据中提取特征。它们能够表征非自主性提示，原因在于：
    - **全连接层**：这种层可以学习输入数据的哪些特征最重要，即它可以学习到的权重可以决定哪些感官输入更应该**被突出**或者“**注意到**”。在某种程度上，这反映了一种非自主性注意，因为网络学习到的是无需明确任务意图，而是仅根据数据本身特性的突出部分。
    - **池化层**：最大池化和平均池化是特征提取过程中的简化步骤，它们通过对邻近特征值的下采样来减少计算量。最大池化层只保留最显著的信号（例如，最亮的像素或最强的特征激活），这类似于非自主性注意力中只对最突出的部分作出响应。平均池化层则提供区域特征的概括，也可以捕捉到区域内的显著信号，但是以平均化的方式。
    - 通过这些层，神经网络能够在不直接受到高层任务指令影响的情况下，根据输入数据的强度或特征本身的显著性，分配其内部“**注意力**”。这是模拟人类非自主性注意力在无需特定任务指导下对感官输入自然反应的方式。


- **（2）通俗理解图10.1.3关于注意力中的查询、键、值以及自主性提示和非自主性提示的作用？**
  - 图10.1.3的核心在于解释如何通过自主性提示（查询）与非自主性提示（键）相结合，在大量感官输入（值）中找到关注的信息。
  - 用一个简单的日常例子来通俗理解这一过程：在图书馆找书。
    - 假设作为读者的你进入了一个巨大的图书馆，想找一本关于“深度学习”的书。这里，“深度学习”就是你的查询（自主性提示），这是一个明确的目标或需求，指导你去寻找某种特定的信息。
    - 图书馆中的每本书都有一个索引卡（键），索引卡上记录了书的主题、作者等信息。这些索引卡就像是非自主性提示，因为它们描述了书的特征而不针对任何特定的查询。
    - 书本（值）就是那些实际的信息或感官输入，每本书与一个特定的索引卡相对应。
    - 现在，为了找到你想要的“深度学习”相关书籍，你需要通过你的查询（“深度学习”这个自主性提示）去匹配图书馆的索引卡（非自主性提示，即键）。这个过程就是注意力机制的体现——根据查询与键的匹配度来决定你的注意力该集中在哪些书（值）上。
    - 当你找到几个与“深度学习”相关的索引卡时，你就可以直接去取相应的书籍了。在这个过程中，**查询帮助你在大量的信息中快速定位到你感兴趣的部分，键和值则提供了一个机制，让这种定位能够实现**。

- **(3)解释“是否包含自主性提示是注意力机制与全连接层或池化层的关键区别”**
  - **自主性提示的本质。**
    - **自主性提示（Volitional Cues）** 指的是一种**动态、输入驱动**的机制，它能够根据具体的输入特征，主动地决定当前任务或场景下需要关注的信息。
    - 在注意力机制中，这种自主性提示被称为“**查询（query）**”。查询是由任务的需求、上下文或输入的特定特征动态生成的，**指导**注意力机制去**选择**和**匹配**相关的信息。
  - **注意力机制中的自主性提示**。在注意力机制中，查询（query）是**输入特定**的，且会动态地与键（key）进行**匹配**，从而**选择最相关**的感官输入（value）。这一过程可以概括为以下几个步骤：
    - **查询（Query）**：代表主动提示，表示模型需要什么信息（即任务需求）。
    - **键（Key）**：表示感官输入（sensory inputs，例如中间特征表示）的特征索引，起到“非自主提示”（Non-volitional Cues）的作用。
    - **值（Value）**：真正的感官输入，是注意力机制最终要选择的信息。
    - 通过注意力池化（attention pooling），查询动态地决定与哪些键匹配，从而选择出对应的值。这种动态匹配的过程体现了注意力机制的**自适应性和灵活性**，即它能够根据任务需求（自主性提示）调整模型的关注点。
  - **全连接层与池化层的非自主性**。
    - 与注意力机制相比，全连接层和池化层的操作是**固定的**，不包含“自主性提示”，即无法根据输入的特性动态调整关注点。
    - **全连接层**
      - 全连接层的权重是固定的（经过训练后确定），对所有输入执行相同的线性变换。
      - 其操作是静态的，完全依赖于预定义的权重矩阵，无法根据输入数据的特性动态调整权重或选择重要的信息。
    - **池化层**
      - 池化层（如最大池化或平均池化）的操作规则同样是固定的，例如选择局部区域的最大值或平均值。
      - 池化层并不会根据任务需求或上下文对输入特征进行选择性处理，其只是一种**固定的降维操作**。
  - **注意力机制与全连接层、池化层的区别**。
    - **动态性**：注意力机制能够动态地根据查询（query）选择与任务相关的信息，而全连接层和池化层的操作是静态的。
    - **任务驱动**：注意力机制通过查询（自主性提示）体现了任务驱动的特性，能够根据任务需求选择性地关注输入中的某些部分。全连接层和池化层则没有这样的能力。
    - **输入依赖**：注意力机制的权重分配是输入依赖的，通过查询和键的匹配动态生成权重。而全连接层的权重是预定义的，池化层也没有动态权重分配的能力。
  - **更直观的解释**。可以通过以下类比来理解这个区别：
    - **注意力机制**：就像一个主动寻找信息的人，根据当前的需求（查询）去书架上寻找和目标（键）匹配的书籍（值）。这个过程是动态的，因为需求不同（查询不同），选择的书籍也不同。
    - **全连接层**：相当于一个固定的目录，书架上的每本书都被按照固定的规则处理，不管需求是什么，操作都是一成不变的。
    - **池化层**：就像一个简单的规则，例如“每行书架只选择最左边的书”或“每行书架的书名取平均值”，因此没有根据需求调整选择的能力。

## 10.1.3 注意力的可视化

- 平均池化层可以被视为输入的加权平均值，
其中各输入的权重是一样的。
- 实际上，注意力池化得到的是加权平均的总和值，
其中权重是在给定的查询和不同的键之间计算得出的。

----------
- **说明：如何理解“注意力池化得到的是加权平均的总和值， 其中权重是在给定的查询和不同的键之间计算得出的。”？**
  - 这句话描述了注意力机制如何工作的一个重要方面，可以用一个类比来更加形象地解释：
  - 想象你在一个水果市场买水果。这里的水果摊位（值）是非常多样化的，每个摊位都有苹果、香蕉、橙子等不同水果出售。你的目标（查询）是买一些新鲜又甜的水果。摊位老板根据水果的新鲜程度、甜度等特点（键）来推荐他们的水果。
  - 在这个情境中：
    - **查询**就是你的需求：新鲜又甜的水果。
    - **键**就是水果的特点，例如新鲜程度、甜度。
    - **值**就是实际的水果或者说你最终想要得到的东西。
  - 理解为加权平均的过程就是说，你会根据每个摊位的水果新鲜程度和甜度给予不同的权重，那些最匹配你的查询的摊位会得到更高的权重。比如，一个摊位的水果非常新鲜又非常甜，你可能会给它赋予很高的权重，这样在你考虑从哪个摊位买水果时，这个摊位的影响就更大。
  - 将这个概念应用回注意力机制：
    - 当模型接收到一个查询时，它会计算这个查询与各个键之间的匹配程度。
    - 每个键所对应的值会根据匹配程度被赋予不同的权重。
    - 然后，模型会计算这些加权的值的总和，这就是加权平均值。
    - 这个总和表现了输入中与查询最相关的部分。
  - 在数学上，如果将权重视作一个归一化的分数（它们加起来等于1），那么加权平均的总和值实际上是一个期望值。
  - 在注意力机制的术语中，这种加权和常常是通过一种叫做“softmax函数”的操作来实现权重的归一化，确保所有权重加起来等于1。
  - 通过这种方式，**模型能够集中其注意力在最相关的信息上**，而忽略或者减少对不那么相关信息的关注。

```python
%matplotlib inline
import torch
from d2l import torch as d2l
```

- 为了可视化注意力权重，我们定义了`show_heatmaps`函数。
- 其输入`matrices`的形状是：
  - 要显示的行数
  - 要显示的列数
  - 查询的数目
  - 键的数目


```python
#@save
def show_heatmaps(matrices, xlabel, ylabel, titles=None, figsize=(2.5, 2.5),
                  cmap='Reds'):
    """显示矩阵热图"""
    d2l.use_svg_display()
    num_rows, num_cols = matrices.shape[0], matrices.shape[1]
    fig, axes = d2l.plt.subplots(num_rows, num_cols, figsize=figsize,
                                 sharex=True, sharey=True, squeeze=False)
    for i, (row_axes, row_matrices) in enumerate(zip(axes, matrices)):
        for j, (ax, matrix) in enumerate(zip(row_axes, row_matrices)):
            pcm = ax.imshow(matrix.detach().numpy(), cmap=cmap)
            if i == num_rows - 1:
                ax.set_xlabel(xlabel)
            if j == 0:
                ax.set_ylabel(ylabel)
            if titles:
                ax.set_title(titles[j])
    fig.colorbar(pcm, ax=axes, shrink=0.6);
```

- 下面使用一个简单的例子进行演示。
- 在本例子中，仅当查询和键相同时，注意力权重为1，否则为0。


```python
attention_weights = torch.eye(10).reshape((1, 1, 10, 10))
show_heatmaps(attention_weights, xlabel='Keys', ylabel='Queries')
```

在后面的章节中，我们将经常调用`show_heatmaps`函数来显示注意力权重。

- **说明：注意力热图的具体功能说明**
  - 在这个上下文中，"区域"指的是热图中的某个特定部分或区域，它对应于注意力权重矩阵中的一组特定元素。
  - 在注意力机制的场景中，每个元素通常表示一个特定的查询-键对的注意力权重。
  - 因此，热图中颜色较深的区域代表了对应查询-键对的注意力权重较大，说明模型在处理输入时，将更多的关注度放在了这些查询-键对所对应的特定部分上。
  - 这有助于理解模型在处理数据时关注的重点，从而更好地理解模型的行为和决策过程。
  - **注意力权重热图中深浅不同的区域分布可以揭示模型在处理输入数据时的注意力分布**，即模型在解决特定问题时对输入数据的不同部分的关注程度。这对于理解和解释模型的行为特别重要，具体如下：
    - **找出重要特征**：深色区域对应的是模型关注度较高的部分，即模型认为这些部分对于预测结果更为重要。这对于理解模型认为什么特征对预测结果更重要非常有用。
    - **模型错误的诊断**：如果模型的预测结果错误，可以通过查看注意力权重热图来了解模型是否将注意力集中在了错误的位置。这可以帮助更好地理解模型的错误，并找出改进模型的方法。
    - **解释模型预测**：注意力权重热图可以帮助理解模型做出特定预测的原因。通过查看模型关注了哪些特征，可以了解模型为何做出这样的预测。
    - **检测模型偏见**：如果模型在某些特定的特征上过度关注，可能会导致模型偏见。注意力权重热图可以帮助检测这种情况，并采取措施纠正。

## 小结

* 人类的注意力是有限的、有价值和稀缺的资源。
* **受试者使用非自主性和自主性提示有选择性地引导注意力。前者基于突出性，后者则依赖于意识**。
* 注意力机制与全连接层或者池化层的区别源于增加的自主提示。
* 由于包含了自主性提示，注意力机制与全连接的层或池化层不同。
* 注意力机制通过注意力池化使选择偏向于值（感官输入），其中包含查询（自主性提示）和键（非自主性提示）。键和值是成对的。
* 我们可以可视化查询和键之间的注意力权重。
